```python
import numpy as np
import matplotlib.pyplot as plt
import time
import os
import random as rn
import pandas as pd
```

## Selecionar apenas a matriz superior e remover colunas que sejam sempre 0

```python
df_train_numerical = pd.read_csv('./datasets/numerical_train.csv')
df_teste_numerical = pd.read_csv('./datasets/numerical_teste.csv')
df_train_image = pd.read_csv( './datasets/dataset.csv')
df_test_image = pd.read_csv('./datasets/test_dataset.csv')
```

```python
df_train_image
```

```python
numpy_array = df_train_image.to_numpy()
```

```python
new_array = []
for i in numpy_array:
    arr_2d = np.reshape(i, (90, 90))
    arr_2d = np.triu(arr_2d)
    new_array.append(arr_2d.flatten())
    
train_cleaning = pd.DataFrame(new_array)
```

```python
numpy_array_test = df_test_image.to_numpy()
new_array = []
for i in numpy_array_test:
    arr_2d = np.reshape(i, (90, 90))
    arr_2d = np.triu(arr_2d)
    new_array.append(arr_2d.flatten())
    
cleaningtest = pd.DataFrame(new_array)
```

```python
df_train_image_clean = train_cleaning[[c for c
        in list(train_cleaning)
        if (len(train_cleaning[c].unique()) or len(cleaningtest[c].unique())) > 1]]
```

```python
df_test_image_clean = cleaningtest[[c for c
        in list(cleaningtest)
        if (len(train_cleaning[c].unique()) or len(cleaningtest[c].unique())) > 1]]
```

```python
df_train_numerical
```

### One Hot encoding do Sexo

#### Treino

```python
sex=pd.get_dummies(df_train_numerical['sex'])
```

```python
df_train_numerical=df_train_numerical.drop('sex',axis=1)
```

```python
sex.rename(columns = {0:'Female', 1:'Male'}, inplace = True)
```

```python
df_train_numerical = pd.concat([df_train_numerical,sex],axis=1)
```

#### Test

```python
sex=pd.get_dummies(df_teste_numerical['sex'])
```

```python
df_teste_numerical=df_teste_numerical.drop('sex',axis=1)
```

```python
sex.rename(columns = {0:'Female', 1:'Male'}, inplace = True)
```

```python
df_teste_numerical = pd.concat([df_teste_numerical,sex],axis=1)
```

### Gravação dos datasets para usar na CNN1D

```python
df_train_numerical.to_csv('datasets/numericalDataset_train.csv',index=False)
df_teste_numerical.to_csv('datasets/numericalDataset_teste.csv',index=False)
```

```python
df_train_image_clean.to_csv('datasets/conv1dR_train.csv',index=False)
df_test_image_clean.to_csv('datasets/conv1dR_teste.csv',index=False)
```

### Para usar no MLP

```python
df_train = pd.concat([df_train_numerical,df_train_image_clean],axis=1)
```
